```python
import glob
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Specify the folder path
folder_path = './'

# Get a list of all CSV files in the folder
csv_files_largepulses = glob.glob(os.path.join(folder_path, '*lar*.csv'))
csv_files_regmeasurements = glob.glob(os.path.join(folder_path, '*mea*.csv'))
```

```python
# Loop through the list of CSV files and read each one
plt.figure(dpi=150)

for file in csv_files_largepulses:
    df = pd.read_csv(file, skiprows=3, header=None)
    time = np.array(df.iloc[:,0])
    voltage = np.array(df.iloc[:,1])
    plt.plot(time,voltage)
plt.show() 
```

```python
# Loop through the list of CSV files and read each one
plt.figure(dpi=150)

for file in csv_files_regmeasurements:
    df = pd.read_csv(file, skiprows=3, header=None)
    time = np.array(df.iloc[:,0])
    voltage = np.array(df.iloc[:,1])
    plt.plot(time,voltage)
plt.show() 
```
