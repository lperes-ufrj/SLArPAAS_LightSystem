```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = 'FirstSiPM_ScopeData/2stagespreampboard1chip.csv'
# Read the CSV file into a pandas DataFrame
data = pd.read_csv(csv_file,skiprows=1)
data.head()

# Extract the columns
x_bk = data.iloc[:, 0]*10**6  # the first column is seconds
y_bk = data.iloc[:, 1]  # the second column is Volts (WaveGenerator)
#y2 = data.iloc[:, 2]  # the second column is Volts (SiPM)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x_bk, y_bk, marker='', linestyle='-', label='SiPM')
#plt.plot(x, y2, marker='.', linestyle='--',  label='SiPM')
plt.ylabel('Volts (V)')
plt.xlabel(r'micro seconds ($\mu$s)')
plt.grid()
#plt.yscale('log')
plt.title('Plot of CSV Data Oscilloscope')
plt.legend(loc='upper right', title = f'Max. Amplitude Signal SiPM P2P: {round(y_bk.max()-y_bk.min(),5)} V', ncols=2)
plt.show()
```
