```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
from scipy import integrate

# Specify the folder path
folder_path = './GainMeasurement'

# Get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

hist_c = []
# Loop through the list of CSV files and read each one
for file in csv_files:
    df = pd.read_csv(file, skiprows=3, header=None)
    time = np.array(df.iloc[:,0])
    voltage = np.array(df.iloc[:,1])

    #print(time)
    #print(voltage)
    voltage = voltage[time>0.5e-6]
    time = time[time>0.5e-6]
    voltage = voltage[time<1.25e-6]
    time = time[time<1.25e-6]

    #voltage = voltage*1e3
    #time = time*1e6

    hist_c.append(np.trapz(-voltage,time))
```

```python
print(hist_c)
```

```python
plt.figure(dpi=150)
plt.hist(hist_c, bins = 30, density =1 )
plt.xlabel(r' "Charge" (V $\cdot$ s)')
plt.show()
```
