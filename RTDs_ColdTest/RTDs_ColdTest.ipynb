```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
RTD1 = np.loadtxt('datalog_sensor1.txt', delimiter=',')
RTD2 = np.loadtxt('datalog_sensor2.txt', delimiter=',')
RTD5 = np.loadtxt('datalog_sensor5.txt', delimiter=',')
RTD6 = np.loadtxt('datalog_sensor6.txt', delimiter=',')
RTD7 = np.loadtxt('datalog_sensor7.txt', delimiter=',')

RTDs = [RTD1,RTD2,RTD5,RTD6,RTD7]
```

```python
plt.figure(dpi=150)
plt.plot(RTD1[:,0], RTD1[:,1], label = 'RTD1')
plt.plot(RTD2[:,0], RTD2[:,1], label = 'RTD2')
plt.plot(RTD5[:,0], RTD5[:,1], label = 'RTD5')
plt.plot(RTD6[:,0], RTD6[:,1], label = 'RTD6')
plt.plot(RTD7[:,0], RTD7[:,1], label = 'RTD7')
plt.hlines(-196, 0 ,900, ls= '--', color ='darkgreen', alpha = 0.8, label= r'LN$_2$')
plt.ylabel('Temperature (C)')
plt.xlabel('Time (s)')
plt.legend(loc = 'center right')
plt.show()
```

```python
RTDs_label = [1,2,5,6,7] 

# Create a 3x2 grid of subplots
fig, axes = plt.subplots(5, 1, figsize=(10, 8))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each graph
for i, ax in enumerate(axes):
    ax.plot(RTDs[i][:,0],RTDs[i][:,1], label = f'RTD {RTDs_label[i]}')
    ax.legend(loc = 'lower right')

# Add a single unique x label at the bottom
fig.supxlabel('Time (s)',fontsize = 17)

# Add a single unique y label on the left
fig.supylabel('Temperature (C)')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

```python
InLN2_RTD7 = RTD7[(RTD7[:,0]>30) & (RTD7[:,0]<60)]
InLN2_RTD5 = RTD5[(RTD5[:,0]>100) & (RTD5[:,0]<250)]
InLN2_RTD6 = RTD6[(RTD6[:,0]>50) & (RTD6[:,0]<200)]
InLN2_RTD2 = RTD2[(RTD2[:,0]>100) & (RTD2[:,0]<400)]
InLN2_RTD1 = RTD1[(RTD1[:,0]>450) & (RTD1[:,0]<690)]

InLN2_RTDs = [InLN2_RTD1, InLN2_RTD2, InLN2_RTD5, InLN2_RTD6, InLN2_RTD7]
```

```python
plt.figure(dpi=150)
for id, rtd in enumerate(InLN2_RTDs):
    plt.scatter(-196,np.median(rtd[:,1]),label = RTDs_label[id])
    plt.errorbar(-196, np.median(rtd[:,1]), yerr=rtd[:,1].std(), fmt="o")
plt.ylim([-197,-195])
plt.hlines(-196,-205,-187.5, ls= '--')

plt.legend()
plt.show()
```
